### WINGS의 Integration Daily on the book by Hotel 활용

```python
import pandas as pd
```

```python
yy = ['2003']
mm = ['07','08','09','10','11','12']
```

```python
path = 'data/'
data = pd.DataFrame()

for i in yy:
      for j in mm:
            table = pd.read_excel(path + i + '_' + j+'.xls', header = 1)
            data = pd.concat([data,table])
```

```python
data.head()
```

```python
hotel = ['제주','남원','경주','낙산','드림','도고','가평']
col = data.columns
```

```python
col
```

```python
new_col = ['Date','Day']
for i in hotel:
      for j in range(2, 9):
            ncol = i + col[j]
            new_col.append(ncol)
```

```python
new_col
```

```python
data.columns = new_col
```

```python
data
```

```python
data2003 = data[data['Date']!= 'Total']
data2003
```

```python
#객실운영률 최대 67%로 제한
df['2020-12-18':'2021-02-14'].head(3)
```

```python
#객실운영률 최대 75%로 제한
df['2021-07-19':'2021-08-17'].head(3)
```

```python
#객실운영률 최대 67%로 제한
df['2021-08-18':'2021-09-22'].head(3)
```

```python
#객실운영률 최대 75%로 제한
df['2021-09-23':'2021-10-17'].head(3)
```

```python
def scale67(x):
    return (x*100)/67
```

```python
def scale75(x):
    return (x*100)/75
```

```python
df.loc['2020-12-18':'2021-02-14', '제주Occ%'] = round(df['2020-12-18':'2021-02-14']['제주Occ%'].apply(scale67),1)
df.loc['2021-07-19':'2021-08-17', '제주Occ%'] = round(df['2021-07-19':'2021-08-17']['제주Occ%'].apply(scale75),1)
df.loc['2021-08-18':'2021-09-22', '제주Occ%'] = round(df['2021-08-18':'2021-09-22']['제주Occ%'].apply(scale67),1)
df.loc['2021-09-23':'2021-10-17', '제주Occ%'] = round(df['2021-09-23':'2021-10-17']['제주Occ%'].apply(scale75),1)
```

```python
df.to_excel('data(2003.7_2022.02.21).xlsx', header = True, index=False)
```
