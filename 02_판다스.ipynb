## 기초적인 통계량 계산하기

```python
df.head(14)
```

```python
# 연도별 lifeExp의 평균
df.groupby('year')['lifeExp'].mean()
```

```python
# df를 연도별로 그룹화한 데이터의 자료형
type(df.groupby('year'))
```

```python
# df를 연도별로 그룹화한 데이터가 저장된 메모리의 위치
df.groupby('year')
```

```python
# df를 연도별로 그룹화한 뒤 lifeExp만 추출한 데이터의 자료형
type(df.groupby('year')['lifeExp'])
```

```python
# 연도별로 그룹화한 lifeExp의 평균
df.groupby('year')['lifeExp'].mean()
```

```python
# 연도 그리고 대륙별 평균 lifeExp과 gdpPercap
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean().head(10)
```

```python
type(df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean())
```

```python
#대륙별 국가의 숫자
df.groupby('continent')['country'].nunique()
```

```python
df.groupby('continent')['country'].value_counts()
```
