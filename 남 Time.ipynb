### WINGS의 Integration Daily on the book by Hotel 활용

```python
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
```

```python
data = pd.read_excel('data(2003.7_2022.02.21).xlsx')
```

```python
#data[data['Date']=='2019-05-01'].index
#2019년 5월 1일 이후 최종예약률을 활용
data = data.loc[5783:,]
```

```python
data.head()
```

```python
data.tail()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data.describe()
```

```python
df = data[['Date', '남원Occ%']]
```

```python
df.columns = ['ds', 'y']
```

```python
df
```

```python
model = Prophet()
```

```python
#model.add_country_holidays(country_name='KR')
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=67, freq='1D')
```

```python
pred = model.predict(future)
```

```python
pred = pred[['ds','yhat','yhat_lower','yhat_upper']]
pred[['ds', 'yhat']].tail()
```

```python
pred.to_csv('남원Occ(2022.2.29~).csv')
```
