### Room Type 선택

```python
def room_type_checker() :  
    driver.find_element_by_css_selector("#frmS > div > table > tbody > tr:nth-child(17) > td > div").click()
    
    # page scroll 
    
    driver.switch_to.default_content()

    
    SCROLL_PAUSE_SEC = 0.1

    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(SCROLL_PAUSE_SEC)

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    driver.switch_to.frame(iframe_)
    
    time.sleep(0.1)
    driver.find_element_by_css_selector("#frmS_roomTypeCode_listbox > li:nth-child(2) > input").click()
    
    time.sleep(0.1)    
    driver.find_element_by_css_selector("#frmS_roomTypeCode_listbox > li:nth-child(3) > input").click()
    
    driver.find_element_by_css_selector("#frmS > div > table > tbody > tr:nth-child(17) > td > div").click() 
```

# 경쟁호텔 객실료

RPA를 통해 수집되는 히든클리프, 위호텔의 향후 60일 객실료

```python
keys = os.path.join('data/예약data/add/', 'JEJUHotel' + datetime.today().strftime("%Y%m%d") + '.xlsx')

a = s3Client.get_object(Bucket=bucket_name, Key=keys)
rpa1 = pd.read_excel(io.BytesIO(a['Body'].read()), skiprows=range(4))
```

```python
from pytz import timezone

start_date = datetime.now(timezone('Asia/Seoul'))
end_date = start_date + timedelta(days=59)

daterange = pd.date_range(start_date, end_date)
```

```python
rpa2 = rpa1.copy()
rpa2['예약일'] = datetime.today().strftime("%Y-%m-%d")
rpa2['입실연도'] = daterange.year
rpa2['입실월'] = daterange.month.astype(str)
rpa2['입실일'] = daterange.day.astype(str)
rpa2['퇴실일'] = (daterange + timedelta(days=1)).day
```

```python
rpa3 = rpa2[['예약일', '입실연도', '입실월', '입실일', '퇴실일', '제주 교원 스위트호텔', '제주 신라 호텔', '롯데 호텔 제주',
             '히든크리프 호텔 네이쳐', '위 호텔 제주', '블룸호텔', '중문 베니키아호텔', '유어스 호텔', '더본 호텔']]
```

```python
rpa3.to_excel(download_dir+'/RPA.xlsx', header=True, index=False, encoding='utf-8-sig')

s3_bucket.upload_file(download_dir+'/RPA.xlsx', Key=folder_path+'RPA_'+datetime.today().strftime("%y%m%d")+'.xlsx')

os.remove(download_dir+'/RPA.xlsx')
```
