```python
Total = pd.concat([data2003, data_past, data2021, data2022])
Total
```

```python
df = Total.copy()
df.index = df['Date']
```

```python
#객실운영률 최대 67%로 제한
df['2020-12-18':'2021-02-14'].head(3)
```

```python
#객실운영률 최대 75%로 제한
df['2021-07-19':'2021-08-17'].head(3)
```

```python
#객실운영률 최대 67%로 제한
df['2021-08-18':'2021-09-22'].head(3)
```

```python
#객실운영률 최대 75%로 제한
df['2021-09-23':'2021-10-17'].head(3)
```

```python
def scale67(x):
    return (x*100)/67
```

```python
def scale75(x):
    return (x*100)/75
```

```python
df.loc['2020-12-18':'2021-02-14', '제주Occ%'] = round(df['2020-12-18':'2021-02-14']['제주Occ%'].apply(scale67),1)
df.loc['2021-07-19':'2021-08-17', '제주Occ%'] = round(df['2021-07-19':'2021-08-17']['제주Occ%'].apply(scale75),1)
df.loc['2021-08-18':'2021-09-22', '제주Occ%'] = round(df['2021-08-18':'2021-09-22']['제주Occ%'].apply(scale67),1)
df.loc['2021-09-23':'2021-10-17', '제주Occ%'] = round(df['2021-09-23':'2021-10-17']['제주Occ%'].apply(scale75),1)
```

```python
data = df
```

```python
#2003년 7월 15일 이후 최종예약률을 활용
data = data.iloc[14:,]
```

```python
data.head()
```

```python
data.tail()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data.describe()
```

```python
df = data[['Date', '제주Occ%']]
```

```python
df.columns = ['ds', 'y']
```

```python
df
```

```python
df = df.copy()
df['y'] = np.where(df['y']>100, 100, df['y'])
```

```python
model = Prophet()
#model = Prophet(yearly_seasonality=True, daily_seasonality=True)
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=60, freq='1D')
```

```python
pred = model.predict(future)
```

```python
pred = pred[['ds','yhat','yhat_lower','yhat_upper']]
pred[['ds', 'yhat']].tail()
```

```python
from datetime import datetime
today = datetime.today().strftime("%Y.%m.%d")

pred.to_csv('제주Occ(' + today + '~).csv')
```
